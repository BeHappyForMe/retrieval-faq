```python
## 查看title_len和 question_len.
title_que_diff['title_len'] = title_que_diff.apply(lambda x: len(x.title), axis=1)
title_que_diff['question_len'] = title_que_diff.apply(lambda x: len(str(x.question)) if str(x.question) else 0, axis=1)
title_que_diff
```

```python
## 计算title和quesiton的文本相似度，
## 对于文本相似度高的，取长度较长的作为best_title, 对于文本相似度较低的 将title和question拼接到一起作为best_title
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
from scipy.linalg import norm

def tf_similarity(s1, s2):
    def add_space(s):
        return ' '.join(list(s))
    
    # 将字中间加入空格
    s1, s2 = add_space(s1), add_space(s2)
    # 转化为TF矩阵
    cv = CountVectorizer(tokenizer=lambda s: s.split())
    corpus = [s1, s2]
    vectors = cv.fit_transform(corpus).toarray()
    # 计算TF系数
    return np.dot(vectors[0], vectors[1]) / (norm(vectors[0]) * norm(vectors[1]))


s1 = '你在干啥呢'
s2 = '你在干什么呢'
print(tf_similarity(s1, s2))
```

```python
#计算question和title的文本相似度
title_que_diff['similarity'] = title_que_diff.apply(lambda x: tf_similarity(x.title, x.question), axis=1)
```

```python
## 文本相似度高的，根据下面显示，我们可以直接取长度较长的作为 best_title (此处0.4主要是通过观察数据得来的，也可以调整为0.4或者0.35之类的)
title_que_diff[title_que_diff.similarity > 0.4]#[title_que_diff.similarity >0.3]
```

```python
## 文本相似度低的，根据下面显示，我们将title和question 拼接起来，作为best_title
title_que_diff[title_que_diff.similarity <=0.4]
```

### best_title
1. 对于question 为空的，直接取title作为best_title
2. 对于quesiton 与 title相同的，直接取title作为best_title
3. 对于question 不为空，且与title不同的，按照上述方法，计算question和title的问题相似度，若相似度大于0.4直接取长度较长的作为best_title, 若相似度小于0.4，则将title与quesiton 拼接起来作为best_title.

```python
def getBestTitle(title, question):
    if question.lower()=='nan':
        return title
    if str(question) == str(title):
        return title
    else:
        sim = tf_similarity(title, question)
        if sim < 0.4:
            return str(title) + str(' ') + str(question)
        else:
            return title if len(title) > len(question) else question
        
```

```python
data['best_title'] = data.apply(lambda x: getBestTitle(x.title, str(x.question)), axis=1)
```

### 获取wrong answer (1个)
#### 方法1
- 如果该问题有对应wrong answer 则直接返回wrong answer
- 如果该问题没有对应的wrong answer 则从所有wrong answer中随机选择一个wrong answer 作为该问题的wrong answer
（此处wrong answer 有些会很离谱）
#### 方法2
- 如果该问题有对应wrong answer 则直接返回wrong answer
- 如果该问题没有对应的wrong answer 则从所有right+wrong answer中 随机选择一个非该问题的answer 作为该问题的wrong answer
（这里可以说明wrong answer有些是有意义的，只是不是该问题的答案，但是有些也可可能是很离谱的）
#### 方法3
- 如果该问题有对应wrong answer 则直接返回wrong answer
- 如果该问题没有对应的wrong answer 则从所有right answer中 随机选择一个非该问题的answer 作为该问题的wrong answer
（这里至少可以说明答案是有意义的，只是不是该问题的答案）

### 获取wrong answers(4个)
- 如果该问题对应的wrong answer 有4个或者大于4个，则取其中4个作为wrong answer
- 如果该问题对应的wrong answer 不够4个，则从所有answer中随机选择非该问题的answer作为该问题的wrong answer. 补够4个即可。

### 获取wrong answer (1个) 方法1

```python
## 获取wrong answer 1
## 在new_data里面去查找title 对应的wrong answer，如果有best_item 对应的is_best=0把reply放到这里
## 如果不存在则随机选一个答案放到这里作为wrong answer
import random
negative_ans = new_data[new_data.is_best==0]

def getWrongAns(title,method=1):
    if title in negative_ans.best_title:
        return negative_ans[negative_ans.best_title==title].iloc[0].reply
    else:
        if method == 1:
            return negative_ans.iloc[random.randint(0,negative_ans.shape[0])].reply
        elif method==2:
           # tmp_data = new_data[new_data.best_title!=title]
            item = new_data.iloc[random.randint(0,new_data.shape[0])]
            if item.best_title == title:
                item = new_data.iloc[random.randint(0,new_data.shape[0])]
            return item.reply
        elif method==3:
            item = right_answer.iloc[random.randint(0,right_answer.shape[0]-1)]
            if item.best_title == title:
                item = right_answer.iloc[random.randint(0,right_answer.shape[0]-1)]
            return item.reply
```

```python
## 获取wrong answer
## 在new_data里面去查找title 对应的wrong answer，如果有best_item 对应的is_best=0把reply放到这里
## 如果不存在则随机选一个答案放到这里作为wrong answer
import random
import numpy
negative_ans = new_data[new_data.is_best==0]
## 返回的是一个个dataframe,返回的是一个neg df 需要append到right answer df上。
def getMulWrongAns(title, num=4):
    if title in negative_ans.best_title.tolist():
        if negative_ans[negative_ans.best_title==title].shape[0] >=4:
            return np.array(negative_ans[negative_ans.best_title==title].iloc[:4])
        else:
#             print('neg_index < 4')
            neg_df=negative_ans[negative_ans.best_title==title]
            
            add = num - neg_df.shape[0]
            ## 从right answer中随机获取add个answer
            neg_reply = []
            tmp =random.sample(list(range(0,right_answer.shape[0]-1)), add)
            for index in tmp:
                if right_answer.iloc[index].best_title != title:
                    neg_reply.append([title, right_answer.iloc[index].reply, 0])
            while len(neg_reply)!=add:
                index = random.randint(0,right_answer.shape[0]-1)
                if index not in tmp:
                    tmp.append(index)
                    neg_reply.append([title, right_answer.iloc[index].reply, 0])
#             print(neg_reply)
            neg_df = neg_df.append(pd.DataFrame(neg_reply,columns=['best_title', 'reply', 'is_best']),ignore_index=True)
            
            return np.array(neg_df)
    else:
        neg_reply = []
        neg_df = pd.DataFrame(columns=['best_title', 'reply', 'is_best'])
        tmp = random.sample(list(range(0,right_answer.shape[0]-1)), num)
        for index in tmp:
            if right_answer.iloc[index].best_title != title:
                neg_reply.append([title, right_answer.iloc[index].reply, 0])
        while len(neg_reply)!=num:
            index = random.randint(0,right_answer.shape[0]-1)
            if index not in tmp:
                tmp.append(index)
                neg_reply.append([title, right_answer.iloc[index].reply, 0])
        neg_df = neg_df.append(pd.DataFrame(neg_reply,columns=['best_title', 'reply', 'is_best']),ignore_index=True)
#         print('no neg rep')
#         print(neg_df)
        return np.array(neg_df)
```
