```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
from scipy.linalg import norm

def tf_similarity(t1,t2):
    def add_space(t):
        return ' '.join(list(t))
    t1,t2 = add_space(t1),add_space(t2)
    cv = CountVectorizer(tokenizer=lambda x: x.split())
    corpus = [t1,t2]
    vectors = cv.fit_transform(corpus).toarray()
    return np.dot(vectors[0],vectors[1])/(norm(vectors[0])*norm(vectors[1]))
```

# best_title：

    1、question为空的直接取title
    2、question与title不同的，计算两者相似度，大于>0.35的可以直接取长度长的，<0.35的内容互为补充，做拼接处理
 

```python
def getBestTitle(question,title):
    if str(question).lower() == 'nan':
        return title
    elif question == title:
        return title
    elif tf_similarity(question,title) > 0.35:
        return question if len(question)>len(title) else title
    else:
        return str(question) + str(' ') + str(title)
```

```python
best_title = data.apply(lambda row:getBestTitle(row['question'],row['title']),axis=1)
```

```python
data['best_title'] = best_title
data
```

### 构建的最终数据集里确保每个title都有正确reply
    在这些title里，增加wrong的reply，即negative-sample

```python

wrong_answer=pd.DataFrame(columns=['best_title','reply','is_best'])
wrong_answer['best_title'] = right_answer.best_title
wrong_answer['is_best'] = 0
```

### 获取wrong answer (1个)
#### 方法1
- 如果该问题有对应wrong answer 则直接返回wrong answer
- 如果该问题没有对应的wrong answer 则从所有wrong answer中随机选择一个wrong answer 作为该问题的wrong answer
（此处wrong answer 有些会很离谱）
#### 方法2
- 如果该问题有对应wrong answer 则直接返回wrong answer
- 如果该问题没有对应的wrong answer 则从所有right+wrong answer中 随机选择一个非该问题的answer 作为该问题的wrong answer
（这里可以说明wrong answer有些是有意义的，只是不是该问题的答案，但是有些也可可能是很离谱的）
#### 方法3
- 如果该问题有对应wrong answer 则直接返回wrong answer
- 如果该问题没有对应的wrong answer 则从所有right answer中 随机选择一个非该问题的answer 作为该问题的wrong answer
（这里至少可以说明答案是有意义的，只是不是该问题的答案）


```python
## 获取wrong answer 1
## 在new_data里面去查找title 对应的wrong answer，如果有best_item 对应的is_best=0把reply放到这里
## 如果不存在则随机选一个答案放到这里作为wrong answer
import random
negative_ans = new_data[new_data.is_best==0]

def getWrongAns(title,method=1):
    if title in negative_ans.best_title:
        return negative_ans[negative_ans.best_title==title].iloc[0].reply
    else:
        if method == 1:
            # 从negatives里随机sample
            return negative_ans.iloc[random.randint(0,negative_ans.shape[0])].reply
        elif method==2:
           # 从所有数据里选择一个非该问题的reply
            item = new_data.iloc[random.randint(0,new_data.shape[0])]
            if item.best_title == title:
                item = new_data.iloc[random.randint(0,new_data.shape[0])]
            return item.reply
        elif method==3:
            #从pos_anws中选择非该问题的reply
            item = right_answer.iloc[random.randint(0,right_answer.shape[0]-1)]
            if item.best_title == title:
                item = right_answer.iloc[random.randint(0,right_answer.shape[0]-1)]
            return item.reply
```

```python
wrong_answer['reply'] =wrong_answer.apply(lambda x: getWrongAns(x.best_title, method=3), axis=1)
wrong_answer
```

```python
all_data = right_answer.append(wrong_answer)
all_data.sort_values(by='best_title',inplace=True)
all_data
```

### 参考word2vec中的neg-sample，wrong-answer取4个

```python
## 获取wrong answer
## 在new_data里面去查找title 对应的wrong answer，如果有best_item 对应的is_best=0把reply放到这里
## 如果不存在则随机选一个答案放到这里作为wrong answer
import random
import numpy
negative_ans = new_data[new_data.is_best==0]
## 返回的是一个个dataframe,返回的是一个neg df 需要append到right answer df上。
def getMulWrongAns(title, num=4):
    if title in negative_ans.best_title.tolist():
        if negative_ans[negative_ans.best_title==title].shape[0] >=4:
            return np.array(negative_ans[negative_ans.best_title==title].iloc[:4])
        else:
#             print('neg_index < 4')
            neg_df=negative_ans[negative_ans.best_title==title]
            
            add = num - neg_df.shape[0]
            ## 从right answer中随机获取add个answer
            neg_reply = []
            tmp =random.sample(list(range(0,right_answer.shape[0]-1)), add)
            for index in tmp:
                if right_answer.iloc[index].best_title != title:
                    neg_reply.append([title, right_answer.iloc[index].reply, 0])
            while len(neg_reply)!=add:
                index = random.randint(0,right_answer.shape[0]-1)
                if index not in tmp:
                    tmp.append(index)
                    neg_reply.append([title, right_answer.iloc[index].reply, 0])
#             print(neg_reply)
            neg_df = neg_df.append(pd.DataFrame(neg_reply,columns=['best_title', 'reply', 'is_best']),ignore_index=True)
            
            return np.array(neg_df)
    else:
        neg_reply = []
        neg_df = pd.DataFrame(columns=['best_title', 'reply', 'is_best'])
        tmp = random.sample(list(range(0,right_answer.shape[0]-1)), num)
        for index in tmp:
            if right_answer.iloc[index].best_title != title:
                neg_reply.append([title, right_answer.iloc[index].reply, 0])
        while len(neg_reply)!=num:
            index = random.randint(0,right_answer.shape[0]-1)
            if index not in tmp:
                tmp.append(index)
                neg_reply.append([title, right_answer.iloc[index].reply, 0])
        neg_df = neg_df.append(pd.DataFrame(neg_reply,columns=['best_title', 'reply', 'is_best']),ignore_index=True)
#         print('no neg rep')
#         print(neg_df)
        return np.array(neg_df)
```
